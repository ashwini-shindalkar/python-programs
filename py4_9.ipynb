## Joining dataframes

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3','A4','A5'],
                 'B':['B0','B1','B2','B3','B4','B5'],
                'C':['C0','C1','C2','C3','C4','C5'],
                'D':['D0','D1','D2','D3','D4','D5']})
```

```python
df1
```

```python
df2=pd.DataFrame({'A':['A6','A7','A8','A9','A10','A11'],
                 'B':['B6','B7','B8','B9','B10','B11'],
                'C':['C6','C7','C8','C9','C10','C11'],
                'D':['D6','D7','D8','D9','D10','D11']})
```

```python
df2
```

```python
## when 2 dataframe have same columns and we need to just append the row
df3=pd.concat([df1,df2],axis=0, ignore_index=True)
df3
```

```python
df4=pd.concat([df1,df2],axis=1, ignore_index=False)
df4  ##herer we not ignore data index so it give the same index as once
```

```python
df3.to_csv("concatted_df.csv")
```
